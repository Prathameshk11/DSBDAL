```python
import pandas as pd

df = pd.read_excel("/home/ubuntu/TCOB35/DSBDAL/data.xlsx")
d= df.filter(items=['DM'])
# d = pd.Series(df['DM'])
print(d)
```

```python
import matplotlib.pyplot as plt

fig= plt.figure(figsize=(10,7))
plt.boxplot(d['DM'])

plt.title('Boxplot of DM')
plt.xlabel('DM')
plt.ylabel('Values')
plt.show()
```

```python
plt.hist(d['DM'], bins=24, edgecolor='black')

plt.title('Histogram of DM')
plt.xlabel('DM')
plt.ylabel('Frequency')

plt.show()
```
