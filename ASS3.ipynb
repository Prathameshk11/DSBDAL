```python
import pandas as pd
import numpy as np

csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

col_names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Species']
iris = pd.read_csv(csv_url, names = col_names)

print(iris)
```

```python
mean = iris['Sepal_Length'].mean()
print(mean)
```

```python
median = iris['Sepal_Length'].median()
print(median)
```

```python
mode = iris['Sepal_Length'].mode()
print(mode)
```

```python
mini = iris['Sepal_Length'].min()
print(mini)
```

```python
maxi = iris['Sepal_Length'].max()
print(maxi)
```

```python
# Task 1: Summary Statistics Grouped by Species
summary_stats = iris.groupby('Species').describe().T
print("Summary Statistics Grouped by Species:\n", summary_stats)
```

```python
species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
for species_name in species:
    print(f"\nStatistics for {species_name}:")
    species_data = iris[iris['Species'] == species_name]
    
    numeric_data = species_data.select_dtypes(include=[float, int])  
    
    stats = numeric_data.describe().T  
    median = numeric_data.median()  
    variance = numeric_data.var()  
    
    stats['median'] = median
    stats['var'] = variance
    
    print(stats[['mean','median','var', 'std', 'min', '25%', '50%', '75%', 'max']])
```
