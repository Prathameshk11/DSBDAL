```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
col_names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Species']
iris = pd.read_csv(csv_url, names=col_names)
```

```python
iris_numeric = iris.drop(columns=['Species'])
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris_numeric)
plt.title("Boxplot of Features Before Normalization")
plt.xlabel("Features")
plt.ylabel("Values")
plt.show()
```

```python
# IQR Normalization
Q1 = iris_numeric.quantile(0.25)
Q3 = iris_numeric.quantile(0.75)
IQR = Q3 - Q1
iris_iqr_normalized = (iris_numeric - Q1) / IQR

print("IQR Normalized Data:\n", iris_iqr_normalized.head())
```

```python
# Min-Max Scaling (0 to 1)
scaler = MinMaxScaler()
iris_minmax = pd.DataFrame(scaler.fit_transform(iris_numeric), columns=iris_numeric.columns)

print("\nMin-Max Scaled Data:\n", iris_minmax.head())
```

```python
# Z-score Normalization
scaler = StandardScaler()
iris_zscore = pd.DataFrame(scaler.fit_transform(iris_numeric), columns=iris_numeric.columns)
print("\nZ-score Normalized Data:\n", iris_zscore.head())
```
